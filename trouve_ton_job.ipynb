```python
import pandas as pd
import numpy as np
import re 
import json
import datetime as dt
```

```python
data = pd.read_json('data.json')
df = data.copy()
df.dtypes
```

```python
df.shape
```

```python
df = df.replace('\n','', regex=True)
df['Intitulé du poste'] = [' '.join(item).replace('\n', '').lower() for item in df['Intitulé du poste']]
df['lieu'] = [', '.join(item).replace('\n', '').lower() for item in df['lieu']]
df['competences'] = [' '.join(item).replace('\n', '').replace(' ', ', ').lower() for item in df['competences']]
df['Type de poste'] = [''.join(item).replace('\n', ', ').replace(', , ', '').lower().strip() for item in df['Type de poste']]
# df[['lieu', 'salary']] = df["lieu"].str.split(",", expand = True, n=1)
df
```

```python
y = [' '.join(element.split()) for element in df['Date de publication']]
```

```python
date0 = dt.date(2023, 1, 15)
dates = []
for item in y:
    delai = item.split()
    if delai[-1] == "jours":
        delta = dt.timedelta(days=int(delai[-2]))
        date = date0 - delta
        dates.append(date)
    elif delai[-1] == 'mois':
        delta = dt.timedelta(weeks=4*int(delai[-2]))
        date = date0 - delta
        dates.append(date)
    elif delai[-1] == 'heures':
        delta = dt.timedelta(hours=int(delai[-2]))
        date = date0 - delta
        dates.append(date)
    elif delai[-1] == 'hier':
        date = dt.date(2013, 1, 14)
        dates.append(date)
df['Date de publication'] = dates
df
```

```python
postes = df['Intitulé du poste'].replace(to_replace='-', value='', regex=True).replace(to_replace='[(h/f)]', value='', regex=True)
postes
```

```python
# format des colonnes salaires
s = [item for item in df['lieu']]
pat = re.compile(r'^((\d)[0-9]*)[0-9]\.(\d){2}$')
salaires = []
for item in s:
    test = bool(re.search(r'\d', item))
    if test == True:
        salaires.append(item)
    else:
        salaires.append(np.nan)

salaries = []
for item in salaires:
    if type(item) is float:
        salaries.append(item)
    elif type(item) is str:
        element = item.split(',')[1:]
        print(element)
        if element == []:
            salaries.append(np.nan)
        else:
            salaries.append(element)
df['salaires'] = salaries
df
```

```python
df.to_csv('clean_data.csv')
```
