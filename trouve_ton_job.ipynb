```python
import pandas as pd
import numpy as np
import re 
import json
import datetime as dt
```

```python
data = pd.read_json('data.json')
df = data.copy()
df.dtypes
```

```python
df.shape
```

```python
col1 = [' '.join(item).replace('\n', '').lower() for item in df['Intitulé du poste']]
df['Intitulé du poste'] = col1

col2 = [item.replace('\n', '') for item in df['Date de publication']]
df['Date de publication'] = col2

col3 = [' '.join(item).replace('\n', '').lower() for item in df['lieu']]
df['lieu'] = col3

col4 = [' '.join(item).replace('\n', '').replace(' ', ', ').lower() for item in df['competences']]
df['competences'] = col4

col5 = [''.join(item).replace('\n', ', ').replace(', ,', '').strip().lower() for item in df['Type de poste']]
df['Type de poste'] = col5

df
```

```python
y = [' '.join(element.split()) for element in df['Date de publication']]
```

```python
date0 = dt.date(2023, 1, 15)
dates = []
for item in y:
    date = 0
    delai = item.split()
    if delai[-1] == "jours":
        delta = dt.timedelta(days=int(delai[-2]))
        date = date0 - delta
        dates.append(date)
    elif delai[-1] == 'mois':
        delta = dt.timedelta(weeks=4*int(delai[-2]))
        date = date0 - delta
        dates.append(date)
    elif delai[-1] == 'heures':
        delta = dt.timedelta(hours=int(delai[-2]))
        date = date0 - delta
        dates.append(date)
    elif delai[-1] == 'hier':
        date = dt.date(2013, 1, 14)
        dates.append(date)
df['Date de publication'] = dates
df['Date de publication']
```

```python
postes = df['Intitulé du poste'].replace(to_replace='-', value='', regex=True).replace(to_replace='[(h/f)]', value='', regex=True)
postes
```

```python
# Split single column into several columns use ',' delimiter
df[['lieu', 'salary']] = df["lieu"].str.split(",", expand = True, n=1)# Salaire
# salaire_moyen

# df[['min_salary', 'max_salary']] = df["salary"].str.split("-", expand = True, n=1)
df
```
